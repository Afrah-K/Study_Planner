```python
import pandas as pd
import numpy as np
```

```python
#Explicit Data

explicit_data = pd.read_csv('International students Time management data.csv')
df = pd.DataFrame(explicit_data)
df.head()
```

```python

# Remove column Nationality, Program, English
df.drop(['Nationality', 'Program', 'English', 'Serial'], axis=1, inplace=True)

# Remove ages 31-35, >36, and <18
df = df[~df['Age'].str.contains('31-35|>36|<18')]

# Convert age range into scale
age_mapping = {'18-20': 1, '21-25': 2, '26-30': 3}
df['Age'] = df['Age'].map(age_mapping)

df.head()
```

```python
# Create a new column 'hours_week'
df['hours_week'] = np.random.choice(['<5', '5-10', '10-15', '15-20', '>20'], size=len(df))

# Create a new column 'days_studied_week'
df['days_studied_week'] = np.random.choice(['<2', '3', '4', '5', '>5'], size=len(df))

# Create a new column 'hours_at_time'
df['hours_at_time'] = np.random.choice(['<1', '2', '3', '>4'], size=len(df))

# Create a new column 'time_day_studied'
df['time_day_studied'] = np.random.choice(['morning', 'afternoon', 'evening', 'midnight'], size=len(df))

# Display the updated DataFrame
print(df.head())
```

```python
# Specify the columns to replace values
columns_to_replace1 = ['7', '8', '10', '14', '17']
columns_to_replace2 = ['6', '9', '11', '12', '13', '14', '15', '16']

# Define the mapping of ranks
rank_mapping1 = {'Strong Disagree': 5, 'Disagree': 4, 'Neither': 3, 'Agree': 2, 'Strong Agree': 1}
rank_mapping2 = {'Strong Disagree': 1, 'Disagree': 2, 'Neither': 3, 'Agree': 4, 'Strong Agree': 5}

# Replace the values in the specified columns
df[columns_to_replace1] = df[columns_to_replace1].replace(rank_mapping1)
df[columns_to_replace2] = df[columns_to_replace2].replace(rank_mapping2)

df.head()
```

```python
#S0 means the most attendance, the level of stage will be upgraded by every 3-4 absence, and S4 means the least attendance
attendance_mapping = {'S0': 1, 'S1': 2, 'S2': 3, 'S3': 4, 'S4': 5}

# Map the attendance to unique fields
df['Attendance'] = df['Attendance'].map(attendance_mapping)
```

```python
# Define the mapping of academic ranges to unique fields
academic_mapping = {'<40%': 1, '40%~49%': 2, '50%~59%': 3, '60%~70%': 4, '>70%': 5}

# Map the academic ranges to unique fields and convert to 'Int64' type
df['Academic'] = df['Academic'].map(academic_mapping).astype('Int64')

# Drop rows with NaN values in the 'Academic' column
df = df.dropna(subset=['Academic'])

#modified DataFrame
df.head()
```

```python
# Mapping for 'hours_week'
hours_week_mapping = {'<5': 1, '5-10': 2, '10-15': 3, '15-20': 4, '>20': 5}
df['hours_week'] = df['hours_week'].map(hours_week_mapping)

# Mapping for 'days_studied_week'
days_studied_week_mapping = {'<2': 1, '3': 2, '4': 3, '5': 4, '>5': 5}
df['days_studied_week'] = df['days_studied_week'].map(days_studied_week_mapping)

# Mapping for 'hours_at_time'
hours_at_time_mapping = {'<1': 1, '2': 2, '3': 3, '>4': 4}
df['hours_at_time'] = df['hours_at_time'].map(hours_at_time_mapping)

# Mapping for 'time_day_studied'
time_day_studied_mapping = {'morning': 1, 'afternoon': 2, 'evening': 3, 'midnight': 4}
df['time_day_studied'] = df['time_day_studied'].map(time_day_studied_mapping)

# Display the updated DataFrame
print(df.head())
```

```python
#save as final preprocessed file
df.to_csv('final_data.csv')
```
